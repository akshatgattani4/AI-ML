```python
q1 = data.TotalWorkingYears.quantile(0.25)
q3 = data.TotalWorkingYears.quantile(0.75)
IQR = q3-q1
upper_limit = q3+1.5*IQR
```

```python
data = data[data.TotalWorkingYears < upper_limit]
```

```python
sns.boxplot(data.TotalWorkingYears)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data.Attrition = le.fit_transform(data.Attrition)
data.head()
```

```python
data.info()
```

```python
data.BusinessTravel = le.fit_transform(data.BusinessTravel)
data.Department = le.fit_transform(data.Department)
data.Gender = le.fit_transform(data.Gender)
data.JobRole = le.fit_transform(data.JobRole)
```

```python
sns.barplot(x = 'Gender', y='Attrition', data = data, errorbar =None)
```

```python
sns.barplot(x = 'Department', y='Attrition', data = data, errorbar =None)
```

```python
plt.subplots(figsize = (20,10))
sns.heatmap(data.corr())
```

```python
data.drop(['Age', 'YearsInCurrentRole', 'YearsAtCompany', 'StockOptionLevel', 'TotalWorkingYears', 'JobLevel', 'EmployeeCount','StandardHours' ,'EmployeeNumber'], axis = 1)
```

```python
for col in data.columns:
  if data[col].dtype == np.number:
    continue
  else:
    data[col] = le.fit_transform(data[col])
```

```python
data.head()
```

```python
x = data.drop(['Attrition'], axis = 1)
y = data['Attrition']
```

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
```

```python
x_scaled = pd.DataFrame(ms.fit_transform(x))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
lgr = LogisticRegression()
```

```python
lgr.fit(x_train, y_train)
```

```python
y_pred = lgr.predict(x_test)
pred = pd.DataFrame({"Actual" : y_test, "Predicted" : y_pred})
pred
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, confusion_matrix, accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn import metrics
```

```python
metrics.r2_score(y_test, y_pred)
```

```python
prob = lgr.predict_proba(x_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, prob)
plt.plot(fpr, tpr)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10, criterion = 'entropy', random_state=0)
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
pred = pd.DataFrame({"Actual" : y_test, "Predicted" : y_pred})
pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
metrics.r2_score(y_test, y_pred)
```

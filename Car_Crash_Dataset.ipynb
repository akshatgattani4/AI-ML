###1. Importing libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

###2.Importing dataset

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset("car_crashes")
```

###3. Analysing data

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.corr()
```

```python
sns.heatmap(data.corr(), annot = True)
```

This heatmap shows that alcohol has a direct relation to accidents.

###4. Visualising data

```python
plt.scatter(x = 'total', y = 'alcohol', data = data)
```

```python
plt.scatter(x = 'total', y = 'no_previous', data = data)
```

```python
plt.scatter(x = 'total', y = 'not_distracted', data = data)
```

```python
plt.scatter(x = 'total', y = 'speeding', data = data)
```

```python
sns.distplot(x = data['total'])
```

```python
sns.distplot(x = data['alcohol'])
```

```python
plt.subplots(figsize=(12,16))
df = data.sort_values(by='total', ascending=False)
sns.barplot(data=df,x='total',y='abbrev')
```

This shows that states - SC,ND,WV have the most number of fatal crashes while MD,CA,OR,IL have the least.

```python
plt.subplots(figsize=(12,16))
df = data.sort_values(by='alcohol', ascending=False)
sns.barplot(data=df,x='alcohol',y='abbrev')
```

ND,SC,MT states have most number of incidents under alcohol influence while UT and DC have the least.

```python
plt.subplots(figsize=(12,16))
df = data.sort_values(by='speeding', ascending=False)
sns.barplot(data=df,x='speeding',y='abbrev')
```

HI,PA,SC have most accidents due to speeding while NJ,MA,NE,DC have the least.
